```python
#librerias y entornos a usar
%pylab inline
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.dpi'] =150 # Tama√±o de la imagen
```

```python
N = 512
a=1
patron_simul= np.zeros([a*N,a*N])
                
for i in range(0,N): #eje y
    for j in range(0,N): #eje x
        patron_simul[i,j]= np.cos(2*(i+j)*np.pi/128)+np.cos(2*(i+j)*np.pi/64)
```

```python
plt.imshow((patron_simul),origin='lower',cmap='seismic')             
plt.colorbar()
plt.ylabel('y', fontsize = 10)
plt.xlabel('x', fontsize = 10)
plt.tick_params(axis = 'both', length=5.5, width=0.85, labelsize=10)
plt.xlim([0,N])
plt.ylim([0,N])
#plt.savefig('simul2-punto1.pdf')
plt.show()
```

```python
FT1=np.fft.fft2(patron_simul)
FT2=np.abs(np.fft.fftshift(FT1))
FT=FT2/FT2.max()
```

```python
FT_aperture=FT
result = numpy.where(FT_aperture == np.amax(FT_aperture))
print('Tuple of arrays returned : ', result)
print('List of coordinates of maximum value in Numpy array : ')
# zip the 2 arrays to get the exact coordinates
listOfCordinates = list(zip(result[0], result[1]))
# travese over the list of cordinates
for cord in listOfCordinates:
    print(cord)
```
